### Step 3. Assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url,sep="\t")
print(chipo.shape)
chipo.info()
```

### Step 4. How many products cost more than $10.00?

```python
def to_float(data):
    return data[1:]

chipo["item_price"] = chipo["item_price"].apply(to_float)
chipo["item_price"] = chipo["item_price"].astype("float64")

chipo.item_price.dtype
```

```python
# creating a new column that shows the price of a single item
chipo["product_price"] = chipo['item_price']/chipo["quantity"]
chipo['product_price'].head()
```

```python
chipo1=chipo.drop_duplicates(['item_name','choice_description'])
chipo1 = chipo1.loc[chipo1['product_price']>10,["item_name","choice_description","product_price"]].reset_index(drop=True)
len(chipo1)
```
